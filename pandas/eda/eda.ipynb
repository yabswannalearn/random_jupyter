# EDA

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('world_population.csv')
df
```

### making them two decimal place

```python
pd.set_option('display.float.format', lambda x: '%.2f' % x)
```

### high level overview

```python
df.describe()
```

### checkng the sum counts

```python
df.isnull().sum()
```

### checking for unique values

```python
df.nunique()
```

### sorting columns to see insights

```python
df.sort_values(by='World Population Percentage', ascending=False).head(10)
```

### checking the correlation

```python
df.corr(numeric_only=True)
```

### using a heatmap to visualize the correlation

```python
sns.heatmap(df.corr(numeric_only=True), annot=True)

plt.rcParams['figure.figsize'] = (23,10) ### setting the figsize for all the charts
plt.show()
```

### do this all the time
### group columns
### sorting values by 2022 pop column

```python
df.groupby('Continent').mean(numeric_only=True).sort_values(by='2022 Population', ascending=False)

```

### checking all the country that has oceania continent

```python
df[df['Continent'].str.contains('Oceania')]
```

### UY PHILIPPINES!!!
