# Pandas Visualization

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
print(plt.style.available)
plt.style.use('_mpl-gallery-nogrid')
```

```python
df = pd.read_csv(r'C:\jupyter\pandas\visualization\Ice Cream Ratings.csv')
df = df.set_index('Date')
df
```

```python
df.plot(kind='bar', title='Ice Cream Rating')
```

```python
df.plot.barh(stacked=True)
```

```python
df.plot.scatter(x='Flavor Rating', y='Texture Rating',s = 1000, c='Flavor Rating')
```

```python
df.plot.hist(bins=5)
```

```python
df.boxplot()
```

```python
df.plot.area(figsize=(10,5))
```

```python
df.plot.pie(y='Texture Rating', figsize = (10,7))
```
