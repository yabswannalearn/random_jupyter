# Data cleaning

```python
import pandas as pd
```

```python
df = pd.read_excel("Customer Call List.xlsx")
```

```python
df
```

### dropping duplicates

```python
df = df.drop_duplicates()
```

### dropping columns

```python
df = df.drop(columns='Not_Useful_Column')
```

### removing dirty data on last name column using strip
### u can use lstrip and rstrip, or just strip
### u can also use the '123(then the things you want to remove)'

```python
# df['Last_Name'] = df['Last_Name'].str.strip('/')
# df['Last_Name'] = df['Last_Name'].str.strip('...')
# df['Last_Name'] = df['Last_Name'].str.strip('_')
df['Last_Name'] = df['Last_Name'].str.strip('123/.')

df
```

### removing all the characters that are not an alphabet and number using regex
### alternative for digits only is r'\D'
### using lambda to make the values into a string and putting a dash '-' in between the values
### replacing all the NA to a blank

```python
# df['Phone_Number'] = df['Phone_Number'].str.replace('[^a-zA-Z0-9]', '', regex=True) # step 1

# df['Phone_Number'] = df['Phone_Number'].apply(lambda x: str(x)) # making the rows a string first to use the lambda below # step 2
# df['Phone_Number'] =  df['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10]) # adding a slash inside the values # step 3

df['Phone_Number'] = df['Phone_Number'].str.replace('nan--', '') # removing the NA #step 4
df['Phone_Number'] = df['Phone_Number'].str.replace('Na--', '') # removing the NA

df
```

### splitting the values with a comma, checking 2 commas using n=2
### expand=True means you will expand the table into 3 columns because you separate by 2 commas 

```python
df[['Street_Address', 'State', 'Zip Code']] = df['Address'].str.split(',', n=2, expand=True)
df
```

### replacing yes with y and no with n

```python
df['Paying Customer'] =  df['Paying Customer'].str.replace('Yes', 'Y')
df['Paying Customer'] =  df['Paying Customer'].str.replace('No', 'N')

df

df['Do_Not_Contact'] =  df['Do_Not_Contact'].str.replace('Yes', 'Y')
df['Do_Not_Contact'] =  df['Do_Not_Contact'].str.replace('No', 'N')
df
```

### changing all null values into a blank

```python
# df = df.replace('N/a', '')
# df = df.replace('NaN', '')

df = df.fillna('')
df
```

### using a for loop and dropping everything that has Y in the 'Do_Not_Contact' Column

```python
for x in df.index:
    if df.loc[x, 'Do_Not_Contact']== 'Y':
        df.drop(x, inplace=True)

df
```

### using the for loop to remove null values in the 'Phone_Number' column

```python
for x in df.index:
    if df.loc[x, 'Phone_Number']== '':
        df.drop(x, inplace=True)

df

# df.dropna(subset='Phone_Number'), inplace=True ## alternative for dropping the null in the phone number col
```
