# MERGE, JOIN, AND CONCATENATE


```python
import pandas as pd
```

```python
df1 = pd.read_csv("LOTR.csv")
df2 = pd.read_csv("LOTR 2.csv")
df1
```

```python
df2
```

### inner = all the same only

```python
df1.merge(df2,on=['FellowshipID','FirstName'], how='inner')
```

### outer = all the same and not same

```python
df1.merge(df2,how='outer')
```

### left = all of the left and merong same sa right table

```python
df1.merge(df2,how='left')
```

### right = kabaliktaran lang ng left

```python
df1.merge(df2,how='right')
```

### cross

### kkuha ng isang row sa left table tapos ccompare sa lahat ng right table

```python
df1.merge(df2,how='cross')
```

# Joins function
### this is good for index joining

```python
df1.join(df2, on='FellowshipID', how='outer', lsuffix='_Left', rsuffix='_Right')
```

### setting the index to fellowship id then joining them

```python
df4 = df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'), lsuffix='_Left', rsuffix='_Right', how='outer')
```

```python
df4
```

# Concatenate
### Pumapatong yung table sa baba

```python
pd.concat([df1,df2], join='outer')
```

### parang merge nalang to, jinojoin gamit columns

```python
pd.concat([df1,df2], join='outer', axis=0)
```
